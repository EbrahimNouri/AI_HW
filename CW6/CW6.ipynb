```python
import numpy as np
import matplotlib.pyplot as plt

temperatures = np.random.randint(15, 36, size=7)

print("دماهای 7 روز:", temperatures)

mean_temp = np.mean(temperatures)
max_temp = np.max(temperatures)
min_temp = np.min(temperatures)

print(f"میانگین دما: {mean_temp:.2f} درجه")
print(f"بیشترین دما: {max_temp} درجه")
print(f"کمترین دما: {min_temp} درجه")

days = np.arange(1, 8)

plt.plot(days, temperatures, linestyle='-', marker='+', color='red')

plt.xlabel('روز')
plt.ylabel('دما (درجه سانتی‌گراد)')
plt.title('تغییرات دمای 7 روز متوالی')

plt.grid(True, alpha=0.3)
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("monthly_sales.csv")
grouped = df.groupby(['month', 'product'])['sales'].mean()
# print(grouped)
table = grouped.unstack()
table
table.plot(kind='bar')

# عنوان و برچسب‌های محور
plt.title("Average Monthly Sales by Product")
plt.xlabel("Month")
plt.ylabel("Average Sales")

plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. خواندن فایل داده
df = pd.read_csv("heart.csv")

# 2. انتخاب فقط ستون‌های عددی
numeric_df = df.select_dtypes(include='number')
print("ستون‌های عددی:", numeric_df.columns.tolist())
print("تعداد ستون عددی:", len(numeric_df.columns))

# 3. محاسبه ماتریس همبستگی
corr_matrix = numeric_df.corr()
print("\nماتریس همبستگی:")
print(corr_matrix)

# 4. رسم heatmap با seaborn
plt.figure(figsize=(10,7))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap - Heart Disease Dataset")
plt.show()

# 5. یافتن بیشترین همبستگی منفی با ستون target
target_corr = corr_matrix["target"].sort_values()
most_negative = target_corr.index[0]
print(f"\nویژگی با بیشترین همبستگی منفی با target: {most_negative}")
print(f"مقدار همبستگی: {target_corr.loc[most_negative]}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("tips.csv")

print(df.columns)

if 'day' in df.columns and 'tip' in df.columns:
    plt.figure(figsize=(8,6))
    sns.boxplot(x="day", y="tip", data=df)
    plt.title("Distribution of Tips by Day of Week")
    plt.xlabel("Day")
    plt.ylabel("Tip Amount ($)")
    plt.show()
else:
    print("Columns 'day' and/or 'tip' not found in dataset.")
```
