```python
import math

A = [1, 3, 5, 7, 9]
B = [2, 4, 6, 8]

mean_A = sum(A) / len(A)
print("mean A:", mean_A)
var_A = map(lambda x: (x - mean_A) ** 2, A)
print("var A:", var_A)
std_A = math.sqrt(var_A)
print("std A:", std_A)

mean_B = sum(B) / len(B)
print("mean B:", mean_B)
var_B = map(lambda x: (x - mean_B) ** 2, B)
print("var B:", var_B)
std_B = math.sqrt(var_B)
print("std B:", std_B)

```
