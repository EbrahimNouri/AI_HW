```python
import pandas as pd
import numpy as np

column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df_iris = pd.read_csv("../HW4/resource/iris.data", header=None, names=column_names)
needed_columns = ['species', 'sepal_length', 'sepal_width']
df_iris = df_iris[needed_columns]

ptest = df_iris.iloc[149].copy()
df_iris.drop(149, inplace=True)
# df_iris.reset_index(drop=True, inplace=True)

print("ptest:")
print(ptest)
```

```python
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))


distances = []
for i, row in df_iris.iterrows():
    dist = euclidean_distance(ptest[['sepal_length', 'sepal_width']],
                              row[['sepal_length', 'sepal_width']])
    distances.append((i, dist, row['species']))

distances_df = pd.DataFrame(distances, columns=['index', 'distance', 'species'])
# distances_df['index'].sort_values(ascending=False)
k = 1
nearest_neighbor = min(distances, key=lambda x: x[1])
print(f"ptest species: {nearest_neighbor[2]} and nearest neighbor species: {ptest['species']}")
print(f"Index {nearest_neighbor[0]}, Distance {nearest_neighbor[1]:.4f}, Class {nearest_neighbor[2]}")

```

```python
k = 3
nearest_neighbors = sorted(distances, key=lambda x: x[1])[:k]

for i, (idx, dist, species) in enumerate(nearest_neighbors):
    print(f"{i+1}. Index {idx}, Distance {dist:.4f}, Class {species}")
```
