```python
import numpy as np
import matplotlib.pyplot as plt


def gaussian(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu) ** 2) / (2 * sigma ** 2))


sigmas = [0.5, 1, 3]
x = np.linspace(-10, 10, 1000)
mu = 0
for sigma in sigmas:
    y = gaussian(x, 0, sigma)
    plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gaussian Distribution')
plt.figure(figsize=(10, 10))
plt.show()

```

```python
import numpy as np

rng = np.random.default_rng(42)
data = rng.normal(loc=50, scale=10, size=500)

partial_sample = rng.choice(data, size=400, replace=False)

full_mean = np.mean(data)
full_var = np.var(data)
full_std = np.std(data)

part_mean = np.mean(partial_sample)
part_var = np.var(partial_sample)
part_std = np.std(partial_sample)

import pandas as pd

df_stats = pd.DataFrame({
    "SMPLE TYPE": ["full data(500)", "partial data(30)"],
    "Mean": [full_mean, part_mean],
    "variance": [full_var, part_var],
    "std": [full_std, part_std]
})

df_stats



```

```python
normal = np.random.normal(loc=-2.0, scale=1.0, size=500)
uniform = np.random.normal(loc=5, scale=1.5, size=500)

conceited = np.concatenate((normal, uniform))
plt.hist(conceited, bins=50)
plt.show()

print("mean:", conceited.mean())
print("var:", conceited.var())
print("std:", conceited.std())

print((-2 + 5)/2)
```
