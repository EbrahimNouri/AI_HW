# START

```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive') # Mount to a path without spaces

# Replace 'your_file.csv' with the actual name of your CSV file
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/diabetes.csv')

# Display the first few rows of the DataFrame
print(len(df))
```

# PART 1

```python
# Fill null values with the mean of each column
df_filled = df.fillna(df.mean()).round(0)

# Display the DataFrame with filled null values
display(df_filled)
```

# PART 2

```python
# Display rows with null values in the original DataFrame
display(df_filled[df_filled.isnull().any(axis=1)])
```

# PART 3

```python
df_filled.head()
```

# PART 4

```python
# Create a new DataFrame with the specified columns
new_df = df_filled[['Glucose', 'BloodPressure', 'BMI', 'Age', 'Insulin', 'Outcome']]

# Display the new DataFrame
display(new_df)
```

# PART 5

```python
# Calculate the correlation matrix for the selected columns
correlation_matrix = new_df.corr()

# Display the correlation matrix
display(correlation_matrix)
```

# PART 6

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

# PART 7

```python
# Select the 'Outcome' column from the correlation matrix
outcome_correlation = correlation_matrix['Outcome']

# Drop the 'Outcome' column from the outcome correlation
outcome_correlation_dropped = outcome_correlation.drop('Outcome')

# Find the feature with the maximum correlation to 'Outcome'
feature_with_highest_correlation = outcome_correlation_dropped.idxmax()

# Find the value of the highest correlation
max_correlation_value_with_outcome = outcome_correlation_dropped.max()

# Display the feature with the highest correlation and its value
print(f"Feature with highest correlation to Outcome: {feature_with_highest_correlation}")
print(f"Maximum correlation value with Outcome: {max_correlation_value_with_outcome}")
```
