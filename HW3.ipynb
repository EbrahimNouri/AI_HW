# PART 1

```python
import numpy as np

# Define the number of cities and days
num_cities = 6
num_days = 30

# Create column names for days (e.g., Day_1, Day_2, ...)
day_columns = [f'Day_{i+1}' for i in range(num_days)]

# Create row names for cities (e.g., City_1, City_2, ...)
# city_rows = [f'City_{i+1}' for i in range(num_cities)]
city_rows = ['Tehran', 'Mashhad', 'Isfahan', 'Tabriz', 'Shiraz', 'Ahvaz']

# Generate random temperatures for 6 cities over 30 days (15 to 40 degrees Celsius)
city_temperatures = np.random.randint(15, 41, size=(num_cities, num_days))

# Generate random humidity for 6 cities over 30 days (20% to 80%)
city_humidity = np.random.randint(20, 81, size=(num_cities, num_days))

# Generate random rainfall for 6 cities over 30 days (0mm to 50mm)
city_rainfall = np.random.randint(0, 51, size=(num_cities, num_days))

print("NumPy arrays for temperatures, humidity, and rainfall have been generated.")
```

# PART 2

```python
import pandas as pd

# Convert the numpy arrays to pandas DataFrames
df_city_temperatures = pd.DataFrame(
    city_temperatures,
    index=city_rows,
    columns=day_columns
)

df_city_humidity = pd.DataFrame(
    city_humidity,
    index=city_rows,
    columns=day_columns
)

df_city_rainfall = pd.DataFrame(
    city_rainfall,
    index=city_rows,
    columns=day_columns
)

print("DataFrame for Temperatures:")
display(df_city_temperatures)

print("\nDataFrame for Humidity:")
display(df_city_humidity)

print("\nDataFrame for Rainfall:")
display(df_city_rainfall)
```

```python
import pandas as pd

# Ensure city_rows, day_columns, df_city_temperatures, df_city_humidity, df_city_rainfall are available from previous steps

# Melt the temperature DataFrame
df_temp_melted = df_city_temperatures.reset_index().melt(
    id_vars='index',
    var_name='Day',
    value_name='Temperature'
)
df_temp_melted = df_temp_melted.rename(columns={'index': 'City'})

# Melt the humidity DataFrame
df_humidity_melted = df_city_humidity.reset_index().melt(
    id_vars='index',
    var_name='Day',
    value_name='Humidity'
)
df_humidity_melted = df_humidity_melted.rename(columns={'index': 'City'})

# Melt the rainfall DataFrame
df_rainfall_melted = df_city_rainfall.reset_index().melt(
    id_vars='index',
    var_name='Day',
    value_name='Rainfall'
)
df_rainfall_melted = df_rainfall_melted.rename(columns={'index': 'City'})

# Merge the melted DataFrames
weather_df = pd.merge(df_temp_melted, df_humidity_melted, on=['City', 'Day'])
weather_df = pd.merge(weather_df, df_rainfall_melted, on=['City', 'Day'])

# Convert 'Day' column from 'Day_X' string to integer X
weather_df['Day'] = weather_df['Day'].str.replace('Day_', '').astype(int)

# Display the first few rows of the combined DataFrame
print("Combined Weather DataFrame:")
display(weather_df)

# Optionally, display information about the DataFrame to check data types and non-null counts
print("\nDataFrame Info:")
weather_df.info()
```

# PART 3

# PART 3 - 1

```python
average_weather_per_city = weather_df.groupby('City')[['Temperature', 'Humidity', 'Rainfall']].mean()

print("Average weather metrics per city:")
display(average_weather_per_city)
```
