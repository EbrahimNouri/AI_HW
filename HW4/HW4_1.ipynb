```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = 'resource/iris.data'

column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df_iris = pd.read_csv(file_path, header=None, names=column_names)

# print(df_iris.head())

print("have null? ", df_iris.isnull().values.any())

df_iris.describe()
needed_column = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
df_iris_without_species = df_iris[needed_column]
df_corr = df_iris_without_species.corr()
df_corr
```

```python
df_iris_without_species ['ratio_petal'] = df_iris_without_species['petal_length'] / df_iris_without_species['petal_width']
df_iris_without_species.head()
# plt.figure(figsize=(10, 10))
# plt.scatter(df_corr['petal_length'], df_corr['petal_width'])
# plt.title('petal length vs petal width')
# plt.xlabel('petal length')
# plt.ylabel('petal width')
# plt.show()

sns.heatmap(df_corr, annot=True, fmt="g")
```
