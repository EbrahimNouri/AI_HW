# PART 1

```python
import numpy as np

num_cities = 6
num_days = 30

day_columns = [f'Day_{i+1}' for i in range(num_days)]

city_rows = ['Tehran', 'Mashhad', 'Isfahan', 'Tabriz', 'Shiraz', 'Ahvaz']

city_temperatures = np.random.randint(15, 41, size=(num_cities, num_days))

city_humidity = np.random.randint(20, 81, size=(num_cities, num_days))

city_rainfall = np.random.randint(0, 51, size=(num_cities, num_days))

print("NumPy arrays for temperatures, humidity, and rainfall have been generated.")
print("city temperatures: " ,city_temperatures)
print("city humidity: ", city_humidity)
print("city rainfall: ", city_rainfall)
```

# PART 2

```python
import pandas as pd

df_city_temperatures = pd.DataFrame(
    city_temperatures,
    index=city_rows,
    columns=day_columns
)

df_city_humidity = pd.DataFrame(
    city_humidity,
    index=city_rows,
    columns=day_columns
)

df_city_rainfall = pd.DataFrame(
    city_rainfall,
    index=city_rows,
    columns=day_columns
)

print("DataFrame for Temperatures:")
display(df_city_temperatures)

print("\nDataFrame for Humidity:")
display(df_city_humidity)

print("\nDataFrame for Rainfall:")
display(df_city_rainfall)
```

```python
import pandas as pd

df_temp_melted = df_city_temperatures.reset_index().melt(
    id_vars='index',
    var_name='Day',
    value_name='Temperature'
)
df_temp_melted = df_temp_melted.rename(columns={'index': 'City'})

df_humidity_melted = df_city_humidity.reset_index().melt(
    id_vars='index',
    var_name='Day',
    value_name='Humidity'
)
df_humidity_melted = df_humidity_melted.rename(columns={'index': 'City'})

df_rainfall_melted = df_city_rainfall.reset_index().melt(
    id_vars='index',
    var_name='Day',
    value_name='Rainfall'
)
df_rainfall_melted = df_rainfall_melted.rename(columns={'index': 'City'})

weather_df = pd.merge(df_temp_melted, df_humidity_melted, on=['City', 'Day'])
weather_df = pd.merge(weather_df, df_rainfall_melted, on=['City', 'Day'])

weather_df['Day'] = weather_df['Day'].str.replace('Day_', '').astype(int)

print("Combined Weather DataFrame:")
display(weather_df)

print("\nDataFrame Info:")
weather_df.info()
```

# PART 3 - 1

```python
average_weather_per_city = weather_df.groupby('City').agg({
    'Temperature': 'mean',
    'Humidity': 'mean',
    'Rainfall': 'mean'
})

print("میانگین دما، رطوبت و بارندگی برای هر شهر:")
display(average_weather_per_city)
```

# PART 3 - 2

```python
warmest_city = average_weather_per_city['Temperature'].idxmax()
warmest_temp = average_weather_per_city['Temperature'].max()

coldest_city = average_weather_per_city['Temperature'].idxmin()
coldest_temp = average_weather_per_city['Temperature'].min()

print(f"Warmest City: {warmest_city} (Average Temperature: {warmest_temp:.2f}°C)")
print(f"Coldest City: {coldest_city} (Average Temperature: {coldest_temp:.2f}°C)")
```

#PART 3 - 3


```python
rainy_days_per_city = weather_df[weather_df['Rainfall'] > 10].groupby('City').size()

print("تعداد روزهای بارانی برای هر شهر (بارندگی بیش از 10 میلی‌متر):")
display(rainy_days_per_city)
```

#PART 4

```python
import matplotlib.pyplot as plt

isfahan_data = weather_df[weather_df['City'] == 'Isfahan']

isfahan_temperature = isfahan_data['Temperature']
isfahan_humidity = isfahan_data['Humidity']

plt.figure(figsize=(10, 6))
plt.scatter(isfahan_temperature, isfahan_humidity, color='blue', alpha=0.7)

plt.title('Temperature - Humidity - Isfahan 30 days')
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')

plt.grid(True)

plt.show()
```
