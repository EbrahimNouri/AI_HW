# START

```python
import pandas as pd

# from google.colab import drive
# drive.mount('/content/drive') # Mount to a path without spaces


# df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/diabetes.csv')
df = pd.read_csv('diabetes.csv')

print(len(df))
```

# PART 1

```python
df_filled = df.fillna(df.mean()).round(0)

display(df_filled)
```

# PART 2

```python
display(df_filled[df_filled.isnull().any(axis=1)])
```

# PART 3

```python
df_filled.head()
```

# PART 4

```python
new_df = df_filled[['Glucose', 'BloodPressure', 'BMI', 'Age', 'Insulin', 'Outcome']]

display(new_df)
```

# PART 5

```python
correlation_matrix = new_df.corr()

display(correlation_matrix)
```

# PART 6

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

# PART 7

```python
outcome_correlation = correlation_matrix['Outcome']

outcome_correlation_dropped = outcome_correlation.drop('Outcome')

feature_with_highest_correlation = outcome_correlation_dropped.idxmax()

max_correlation_value_with_outcome = outcome_correlation_dropped.max()

print(f"Feature with highest correlation to Outcome: {feature_with_highest_correlation}")
print(f"Maximum correlation value with Outcome: {max_correlation_value_with_outcome}")
```
