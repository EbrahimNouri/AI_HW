# A

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from scipy.spatial.distance import euclidean, cityblock, chebyshev, cosine, mahalanobis

data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
# print(len(df.columns))
# df.head()
# df.isnull().sum()
df_numeric = df.select_dtypes("number")
# print(len(df_numeric.columns))
df_numeric.head()

s = pd.Series(data.target)
s
```

# B

```python
malignant = df_numeric[s == 0].head(3)
benign = df_numeric[s == 1].head(3)

samples = pd.concat([malignant, benign])
samples.head()
VI = np.linalg.inv(np.cov(df_numeric.T))

def distances(a, b):
    return {
        "Euclidean": euclidean(a, b),
        "Manhattan": cityblock(a, b),
        "Chebyshev": chebyshev(a, b),
        "Cosine": cosine(a, b),
        "Mahalanobis": mahalanobis(a, b, VI)
    }

results = []
idx = samples.index.tolist()

for i in range(len(idx)):
    for j in range(i+1, len(idx)):
        a = samples.loc[idx[i]].values
        b = samples.loc[idx[j]].values
        results.append([idx[i], idx[j], distances(a, b)])

df_distance = pd.DataFrame(results)
df_distance.head(100)
```
